```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.score(x_test,y_test)
```

```python
Sex=x_test.iloc[0]["Sex"]
Length=x_test.iloc[0]["Length"]
Diameter=x_test.iloc[0]["Diameter"]
Height	=x_test.iloc[0]["Height"]
Whole_weight=x_test.iloc[0]["Whole weight"]
Shucked_weight=x_test.iloc[0]["Shucked weight"]
Viscera_weight=x_test.iloc[0]["Viscera weight"]
Shell_weight=x_test.iloc[0]["Shell weight"]
```

```python
predicted_age=(
lr.coef_[0] *Sex +lr.coef_[1] *Length + lr.coef_[2] * Diameter + lr.coef_[3] * Height +lr.coef_[4] * Whole_weight +lr.coef_[5] * Shucked_weight +lr.coef_[6] * Viscera_weight + lr.coef_[7] * Shell_weight +lr.intercept_)
```
