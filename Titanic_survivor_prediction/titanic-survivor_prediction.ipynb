```python
import matplotlib.pyplot as plt
```

```python
#plot1 : survival by sex
sex_survived=df2.groupby(['Sex','Survived']).size().unstack()
sex_survived.plot(kind='bar',figsize=(8,5),color=['black','green'])
plt.title('survival by sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Survived',labels=['No','Yes'])
plt.show()
```

```python
#plot2: survived by pclass
pclass_survived=df2.groupby(['Pclass','Survived']).size().unstack()
pclass_survived.plot(kind='bar',figsize=(8,5),color=['red','lightgreen'])
plt.title('Survived by passenger class')
plt.xlabel('passenger class')
plt.ylabel('Count')
plt.legend(title='Survived',labels=['No' , 'Yes'])  
plt.show()           
                     
```

```python
#survival by age(Histogram)
plt.figure(figsize=(10,6))
plt.hist(df2[df2['Survived']==0]['Age'],bins=30,alpha=0.6,label='Didnot survive',color='maroon')           
plt.hist(df2[df2['Survived']==1]['Age'],bins=30,alpha=0.6,label='Survived',color='green')

plt.title('Survival by Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(title='Survived')
plt.show()           
```
