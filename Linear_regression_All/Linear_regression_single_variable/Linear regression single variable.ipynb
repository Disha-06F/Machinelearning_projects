```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("Home_prices_sample.csv")
df
```

### To check if the data realization follows a linear regression model

```python
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.prices,color='red',marker='+')
```

### To apply a linear regression model and train it using the fit() function

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.prices)
```

### Use the predict() function to estimate the price of a home in a randomly selected area.

```python
reg.predict([[1500]])
```

### According to the equation \( y = mx + c \), we first find \( m \) using the `coef_` function.

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y=reg.coef_ *1500 + reg.intercept_
y
```
